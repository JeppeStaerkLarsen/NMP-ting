```python
#nim sum game
a='   I   '
b='  III  '
c=' IIIII '
d='IIIIIII'
from collections import Counter
from IPython.display import clear_output

clear_output()


def welcome_message():
    print('Welcome to the Nim Sum game!')
    print('Nim Sum is a game for two players.')
    print("In turn choose a row. Remove at least one, at most all of the I's from that row.")
    print("You can only remove I's from one row in a turn ")
    print('Player to remove the last match looses.')
    print('Good luck!')

def print_board(a,b,c,d):  #input the 4 rows as string containing I's or ' '
    print(a)
    print(b)
    print(c)
    print(d)

def count_Is_rows(a,b,c,d):
    return [a.count('I'),b.count('I'),c.count('I'),d.count('I')]
 
def count_Is_total(a,b,c,d):
    return sum(count_Is_rows(a,b,c,d))

def determine_open_rows(a,b,c,d): #returns as string list to compare with input
    i=0
    ok_rows=[]
    count = count_Is_rows(a,b,c,d)
    while i < len(count):
        if count[i]>0:
            ok_rows.append(i+1) #rows ok to choose
            i+=1
        else:
            i+=1
    return list(map(str,ok_rows))
    

def player_choose_row(player1): 
    #Player
    if player1:
        pl='Player 1, '
    else:
        pl='Player 2, '
               
    choice=''
    # While the choice is not a digit, keep asking for input.
    while choice not in determine_open_rows(a,b,c,d):
        
        # we shouldn't convert here, otherwise we get an error on a wrong input
        choice = input("{}pick a row. Top row is 1, bottom row is 4: ".format(pl))
        
        if choice not in determine_open_rows(a,b,c,d):
            #print_board(a,b,c,d)
            print("Sorry, but you did not choose a valid row.\nTop row is 1, bottom row is 4 and there must be at least one I in the row") 
            
    
    return int(choice)
        
def player_choose_Is(player1,number_in_row): 
    #Player
    if player1:
        pl='Player 1, '
    else:
        pl='Player 2, '
            
    choice=''
    # While the choice is not a digit lower than or equal to number I's , keep asking for input.
    while choice not in list(map(str, list(range(1,number_in_row+1)))):
        
        # we shouldn't convert here, otherwise we get an error on a wrong input
        choice = input("{}pick the number of I's you want to remove: ".format(pl))
        
        if choice not in list(map(str, list(range(1,number_in_row+1)))):
            #print_board(a,b,c,d)
            print("Sorry, but you did not choose a valid number of I's. Choose a number between 1 and {}".format(number_in_row)) 
           

    #Optionally you can clear everything after running the function
    #clear_output()
    
    # We can convert once the while loop above has confirmed we have a digit.
    return int(choice)   

def player_turn(player1,a,b,c,d):
    
    player1 = not player1
    
    clear_output()
    print_board(a,b,c,d)
    
    nrow=player_choose_row(player1)
    
    nIs=player_choose_Is(player1,count_Is_rows(a,b,c,d)[nrow-1])
    
    return (nrow,nIs,player1)

def remove_Is(a,b,c,d,nrow,nIs):
    
    i=0
    x=[a,b,c,d]
    Ilist = [letter for letter in x[nrow-1]]

    for j in x[nrow-1]:
        if j=='I':
            break
        else:
            i+=1
    k=0

    while k < nIs:
        Ilist[i]=' '
        i+=1
        k+=1
    
    x[nrow-1]="".join(Ilist) 
    return x
    
            

#GAME LOGIC

welcome_message()          
n=[0,0,False]

while True:
    n=player_turn(n[2],a,b,c,d)

    q=remove_Is(a,b,c,d,n[0],n[1])

    a=q[0]
    b=q[1]
    c=q[2]
    d=q[3]
    
    if count_Is_total(a,b,c,d)<=1:
        break
        
print_board(a,b,c,d)
print('Game Over')


#count_Is_rows(a,b,c,d)
#count_Is_total(a,b,c,d)
    
   

```
